```python
book = sc.textFile(book_location)
book.collect()
```

```python
# Converting book into words using nltk.word_tokenize and removing comma full stop using filter(lambda x: len(x) > 1) 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
words = book.flatMap(word_tokenize).filter(lambda x: len(x) > 1).map(lambda x: x.lower())
words.collect()
```

```python
# parallelize no to get TypeError: 'PipelinedRDD' object is not iterable
rddWords = sc.parallelize(words.collect())
```

```python
# Removed stop words
stop_words = list(nltk.corpus.stopwords.words('english'))
rddWordsFiltered = rddWords.filter(lambda x: x not in stop_words)
rddWordsFiltered.collect()
```

```python
# Getting key value pairs
wordKVPairs = rddWordsFiltered.map(lambda x: (x, 1))
wordKVPairs.take(10)
```

```python
# Final word count
wordCount = wordKVPairs.reduceByKey(lambda x, y: x+y).sortBy(lambda x: -x[1])
wordCount.collect()
```
